### THREE QUBIT BIT FLIP CODE 

```python
%matplotlib inline
# Importing standard Qiskit libraries and configuring account
from qiskit import QuantumCircuit, execute, Aer, IBMQ, QuantumRegister, ClassicalRegister
from qiskit.compiler import transpile, assemble
from qiskit.tools.jupyter import *
from qiskit.visualization import *
# Loading your IBM Q account(s)
IBMQ.load_accounts()
```

```python
q = QuantumRegister(5)
c = ClassicalRegister(3)
bitflip = QuantumCircuit(q,c)
#encoding
bitflip.h(q[0])
bitflip.cx(q[0],q[1])
bitflip.cx(q[0],q[2])
bitflip.measure(q[0],c[0])
bitflip.measure(q[1],c[1])
bitflip.measure(q[2],c[2])
bitflip.draw()
```

```python
backend = Aer.get_backend('qasm_simulator')
shots = 1024
output = execute(bitflip,backend=backend,shots=shots).result()
outputMeasure = output.get_counts()
print(outputMeasure)
plot_histogram(outputMeasure)
```

```python
#decoding
q = QuantumRegister(5)
c = ClassicalRegister(3)
bitflip = QuantumCircuit(q,c)
#encoding
bitflip.h(q[0])
bitflip.cx(q[0],q[1])
bitflip.cx(q[0],q[2])
#bitflip.measure(q[0],c[0])
#bitflip.measure(q[1],c[1])
#bitflip.measure(q[2],c[2])
bitflip.cx(q[0],q[1])
bitflip.cx(q[0],q[2])
bitflip.ccx(q[2],q[1],q[0])
bitflip.measure(q[0],c[0])
bitflip.draw()
```

```python
backend = Aer.get_backend('qasm_simulator')
shots = 1024
output = execute(bitflip,backend=backend,shots=shots).result()
outputMeasure = output.get_counts()
print(outputMeasure)
plot_histogram(outputMeasure)
```
